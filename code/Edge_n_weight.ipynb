## User-User Edge

```python
ls1 = [x for x in new_user_df["user_id"]]
ls2 = [y for y in new_user_df["friends"]]
ls = []
for u, x in zip(ls1, ls2):
    if u != "#NAME?":
        x = [y.strip(" '") for y in x.strip("[]").split(",")]
        for data in x:
            if data != "":
                # print([u, data])
                ls.append([u, data])  # A pair of user_id and friends
```

```python
freq_table = review_train_df.pivot_table(
    index="user_id", columns="business_id", values="text", aggfunc="count"
)

freq_table.fillna(0, inplace=True)
freq_table
```

```python
set(freq_table.iloc[0].loc[lambda df: df.ne(0)].index)
```

```python
def User2UserWeight(user1, user2, freq_table, eps=1e-4):
    numerator = 0
    user1_business = set(freq_table.loc[user1].loc[lambda df: df.ne(0)].index)
    user2_business = set(freq_table.loc[user2].loc[lambda df: df.ne(0)].index)
    for business in user1_business & user2_business:
        numerator += min(
            freq_table.loc[user1][business], freq_table.loc[user2][business]
        )
    numerator += eps
    denominator = len(user1_business & user2_business) + 1
    return round(numerator / denominator, 4)
```

```python
with open("../cleaned_data/graph_data/User2User_edges.txt", "w") as f:
    for data in tqdm(ls):
        try:
            if data[0] in freq_table.index and data[1] in freq_table.index:
                f.write(str(data[0]))
                f.write(",")
                f.write(str(data[1]))
                f.write(",")
                f.write(str(User2UserWeight(data[0], data[1], freq_table)))
                f.write("\n")  # A pair of user_id and friends
        except:
            pass
```

## User-Poi Edge

```python
import torch
from tqdm import tqdm
from transformers import AutoModelForSequenceClassification, AutoTokenizer

device = (
    "cuda"
    if torch.cuda.is_available()
    else ("mps" if torch.backends.mps.is_available() else "cpu")
)
print(device)

model = AutoModelForSequenceClassification.from_pretrained(
    "nihaldsouza1/yelp-rating-classification",
)

tokenizer = AutoTokenizer.from_pretrained(
    "nihaldsouza1/yelp-rating-classification",
    truncation=True,
    max_length=512,
)
```

```python
review_train_df["text"].values
```

```python
tokenizer
```

```python
model.to(device)
```

```python
res = list(review_train_df["text"].values)

star = []
model.eval()
model.to(device)
with torch.no_grad():
    for i in tqdm(range(len(res))):
        inputs = res[i]
        inputs = tokenizer(inputs, return_tensors="pt", truncation=True)
        inputs.to(device)
        k = torch.argmax(model(**inputs).logits[0]).item() + 1
        star.append(k)
```

```python
temp = review_train_df["stars"]
star = pd.Series(star)
temp.loc[temp.isnull()] = star.loc[temp.isnull()]
review_train_df["new_stars"] = (temp + star) / 2
```

```python
review_train_df
```

```python
def User2PoiWeight(user_id, business_id, freq_table):
    star = review_train_df.loc[
        (review_train_df["user_id"] == user_id)
        & (review_train_df["business_id"] == business_id)
    ]["new_stars"]
    return star
```

```python
review_train_df.pivot_table(
    index="user_id",
    columns="business_id",
    values="new_stars",
    aggfunc="count",
    fill_value=0,
)
with open("../cleaned_data/graph_data/User2Poi_edges.txt", "w") as f:
    for user_id in tqdm(freq_table.index):
        for business_id in freq_table.columns:
            if freq_table.loc[user_id][business_id] >= 1:
                f.write(str(user_id))
                f.write(",")
                f.write(str(business_id))
                f.write(",")
                f.write(str(User2PoiWeight(user_id, business_id, freq_table)))
                f.write("\n")
```

```python
# Generate edges map from 1 PoI to another PoI
Poi2Poi = {}
for business in tqdm(
    time_table.columns,
):
    business_record = time_table[business].sort_values().loc[lambda df: df.notnull()]
    after_visit = []
    for user in tqdm(business_record.index, leave=False):
        timeStamp = business_record.loc[user]
        # user trajectory
        user_trajectory = (
            time_table.loc[user].sort_values().loc[lambda df: df.notnull()]
        )

        user_trajectory = user_trajectory.loc[lambda df: df > timeStamp].index
        if len(user_trajectory) > 0:
            after_visit.append(user_trajectory[0])
    after_visit = list(set(after_visit))
    Poi2Poi[business] = after_visit
```

```python
from geopy.distance import geodesic
```

```python
Poi2Poi_dis = {}
negative_exponent = -7
for business, after_visit in tqdm(Poi2Poi.items()):
    for poi in after_visit:
        business_loc = new_business_df.loc[new_business_df["business_id"] == business][
            ["latitude", "longitude"]
        ].values[0]
        poi_loc = new_business_df.loc[new_business_df["business_id"] == poi][
            ["latitude", "longitude"]
        ].values[0]
        Poi2Poi_dis[(business, poi)] = (
            geodesic(business_loc, poi_loc).miles
        ) ** negative_exponent
    Poi2Poi_dis[(business,)] = sum(
        [Poi2Poi_dis[(business, poi)] for poi in after_visit]
    )
```

```python
Poi2Poi_weight = {}
for business1 in time_table.columns:
    for business2 in Poi2Poi[business1]:
        tmp = 0  # sum of weight
        for user in time_table.loc[lambda df: df[business1].notnull()].index:
            user_trajectory = (
                time_table.loc[user].sort_values().loc[lambda df: df.notnull()]
            )
            for idx, poi in enumerate(user_trajectory.index):
                if (
                    user_trajectory.index[idx] == business1
                    and user_trajectory.index[idx + 1] == business2
                    and idx + 1 < len(user_trajectory)
                ):
                    # Calculate distance between two PoI
                    temp += (
                        Poi2Poi_dis[(business1, business2)] / Poi2Poi_dis[(business1,)]
                    )
        Poi2Poi_weight[(business1, business2)] = temp
```

```python
with open("../cleaned_data/graph_data/Poi2Poi_edges.txt", "w") as f:
    for business1 in tqdm(time_table.columns):
        for business2 in Poi2Poi[business1]:
            f.write(str(business1))
            f.write(",")
            f.write(str(business2))
            f.write(",")
            f.write(str(Poi2Poi_weight[(business1, business2)]))
            f.write("\n")
```
