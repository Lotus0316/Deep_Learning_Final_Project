```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
model = AutoModelForSequenceClassification.from_pretrained("nihaldsouza1/yelp-rating-classification")

tokenizer = AutoTokenizer.from_pretrained("nihaldsouza1/yelp-rating-classification")
```

```python
from tqdm import tqdm
import torch
res=list(df['text'].values)

star=[]

for i in tqdm(range(len(res))):
    inputs=res[i]
    inputs = tokenizer(inputs, return_tensors="pt")
    k=torch.argmax(model(**inputs).logits[0]).item()+1
    star.append(k)
```

```python
df['star']=star
```

```python
df['new_star']=star
```

```python
df.to_csv('res2.csv')
```
